```python
import getpass,imaplib,email,os
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
def bag_of_words_similarity(subject):
    bag_of_words = CountVectorizer().fit_transform([subject.lower(), 'Thank you for applying'])
    bag_of_words_similarity = bag_of_words[0] * bag_of_words[1].T
    return (bag_of_words_similarity.toarray()[0][0]/4)*100
```

```python
userClient = input("Enter your email id ")
password  = getpass.getpass("Enter your password ")
```

```python
user = imaplib.IMAP4_SSL("imap.gmail.com") 
user.login(userClient, password) 
```

```python
user.select('Inbox')
```

```python
resp, items = user.search(None, 'ALL') 
items = items[0].split()
from_mail = [] 
date = [] 
body = [] 
mails = {} 
mail_cnt = 0
for emailid in items[::-1]: 
    resp, data = user.fetch(emailid, "(RFC822)")
    for response_part in data:
        if isinstance(response_part, tuple):
            mail = email.message_from_string(list(response_part)[1].decode("utf-8","ignore"))
            if bag_of_words_similarity(mail['subject']) == 100.0: 
                r, d = user.fetch(emailid, "(UID BODY[TEXT])")
                mail_cnt += 1 
                print("Email from scraped with subject-> " + mail['subject']) 
                from_mail.append(mail['from']) 
                date.append(mail['date']) 
                body.append(email.message_from_string(d[0][1].decode("utf-8"))) 
                
mails['from'] = from_mail
mails['datetime'] = date
mails['body'] = [str(x) for x in body]
mails = pd.DataFrame(mails)
print(str(mail_cnt) + ' relevant email(s) is/are found and scraped')
```
